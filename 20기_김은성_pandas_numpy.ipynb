#### 중요하다고 생각되는 부분만 추출

```python
df=data[['quality','residual sugar','pH','alcohol','type']]
df.head(5)
```

#### groupby를 통해 quality로 묶고 통계값(min,mean,max) 확인

```python
df.groupby(by='quality' ).min()
```

```python
df.groupby(by='quality').mean()
```

```python
df.groupby(by='quality').max()
```

#### groupby를 통해 quality/type로 묶고 개수 세기

```python
df.groupby(by= ['quality', 'type']).count()
```

#### 알코올 함량을 기준으로 오름차순 정리 (20개)

```python
df.sort_values(by='alcohol', ascending=False ).head(20)
```

#### 알코올 함량을 기준으로 내림차순 정리 (20개)

```python
df.sort_values(by='alcohol', ascending=True ).head(20)
```

#### quailty가 7 이상인 데이터 추출 (20개)

```python
df[df['quality'] >=7].head(20) #quality 높은순 20개? or 그냥 인덱스 순으로 20개?
```

#### alcohol을 기준으로 10도가 넘으면 'strong' 아니면 'weak'으로 표시 (lambda 쓰세요)

```python
df['degree'] = df['alcohol'].apply(lambda x : 'strong' if x > 10 else 'weak')
df.head()
```

#### merge 실습을 위해 데이터를 인위적으로 나눠볼게요

```python
df_x = data[['quality', 'fixed acidity', 'citric acid']]
```

```python
df_y = data[['quality', 'volatile acidity', 'chlorides', 'alcohol']]
```

#### quality를 기준으로 inner join

```python
df_x.merge(df_y, how = 'inner', on='quality') #오래 걸려서 실행은 생략
```

#### quality를 기준으로 outer join

```python
df_x.merge(df_y, how = 'outer', on='quality') #오래 걸려서 실행은 생략
```
